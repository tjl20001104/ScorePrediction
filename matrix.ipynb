# Matrix

Here is a rating dataset of [Book Recommendation](https://www.kaggle.com/datasets/arashnic/book-recommendation-dataset).

The book information (Books.csv) and user information (Users.csv) in the website can be used in this project.

### Data Detail

Number of users:    **95,513** (row)

Number of item:     **322,101** (col)

Number of ratings:  **1,048,575**

### Data example

|User-ID|ISBN|Book-Rating|
|----|----|----|
|276725|034545104X|0|
|276726|155061224|5|

### Task

Rating prediction. Use the data in `train_ratings.csv` to construct a model. Then predict the ratings of (user, book) pair in `test_rating.csv`. Finally, the RMSE of the predicted rating and the ground truth rating is calculated

### Requirement

1. Use RMSE of predicted rating as metric. Other reasonable metrics are welcome.

2. We encourge you to implement 2 methods. 1) The algorithm in this course. 2) Use deep models and combines the techniques of this course.

### Data split

The train-test is split in `train_ratings.csv` and `test_ratings.csv`. Note that the information in test set cannot be used in training time.

**train_ratings.csv**

Number of users:    **95,513** in **95,513**

Number of item:     **301,865** in **322,101**

Number of ratings:  **943,715** in **1,048,575**


**test_ratings.csv**

Number of users:    **1,000** in **95,513** 

Number of item:     **66,803** in **322,101**

Number of ratings:  **104,856** in **1,048,575**

### Example

A simple `example.py` predict all the rating to 5 and calculate the RMSE.
